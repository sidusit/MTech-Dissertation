```
def text_to_num(text):
        if text[-1] in d:
            num, magnitude = text[:-1], text[-1]
            return Decimal(num) * 10 ** d[magnitude]
        else:
            return Decimal(text)
```

# EDA and Data Cleaning of Economic data

```
# Defining a generic function to convert monthly / quarterly data to daily time series
def returnTimeSeries(stock_ref_data, dateCol_stockData, coarse_data, dateCol_coarseData, reqCols):
  return_matrix = []
  i = 0
  current = coarse_data.loc[i,dateCol_coarseData]
  next = coarse_data.loc[i+1,dateCol_coarseData]
  for d in stock_ref_data[dateCol_stockData].values:
    # print(d, current, next)
    if (d>=current and d<next):
      # print('a')
      tempList = [d]
      tempList.extend(coarse_data.loc[i,reqCols].values)
      return_matrix.append(tempList)
    else:
      if i<coarse_data.shape[0]-2:
        i = i + 1
        # print('b')
        current = coarse_data.loc[i,dateCol_coarseData]
        next = coarse_data.loc[i+1,dateCol_coarseData]
        tempList = [d]
        tempList.extend(coarse_data.loc[i,reqCols].values)
        return_matrix.append(tempList)
      elif i<coarse_data.shape[0]-1:
        # print('c')
        i = i + 1
        tempList = [d]
        tempList.extend(coarse_data.loc[i,reqCols].values)
        return_matrix.append(tempList)
      else:
        # print('d')
        tempList = [d]
        tempList.extend(coarse_data.loc[i,reqCols].values)
        return_matrix.append(tempList)
  return return_matrix
```
